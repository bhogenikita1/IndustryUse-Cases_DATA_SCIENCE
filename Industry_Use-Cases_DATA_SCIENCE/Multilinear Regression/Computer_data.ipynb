```python
from sklearn.preprocessing import PolynomialFeatures

# Create polynomial features (degree = 2)
poly = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly.fit_transform(X)
```

```python
# Split the dataset
X_train_poly, X_test_poly, y_train, y_test = train_test_split(X_poly, y, test_size=0.2, random_state=42)
```

```python
# Train the polynomial regression model
model_poly = LinearRegression()
model_poly.fit(X_train_poly, y_train)
```

```python
# Make predictions
y_pred_poly = model_poly.predict(X_test_poly)

```

```python
#Evaluate performance
mse_poly = mean_squared_error(y_test, y_pred_poly)
r2_poly = r2_score(y_test, y_pred_poly)
```

```python
mse_poly, r2_poly
```

```python
# Test only degrees 1, 2, and 3 to prevent memory issues
degrees = [1, 2, 3]
mse_values = []
r2_values = []

for d in degrees:
    poly = PolynomialFeatures(degree=d, include_bias=False)
    X_poly = poly.fit_transform(X)

    # Split dataset
    X_train_poly, X_test_poly, y_train, y_test = train_test_split(X_poly, y, test_size=0.2, random_state=42)

    # Train model
    model_poly = LinearRegression()
    model_poly.fit(X_train_poly, y_train)

    # Predictions
    y_pred_poly = model_poly.predict(X_test_poly)

    # Evaluate
    mse_values.append(mean_squared_error(y_test, y_pred_poly))
    r2_values.append(r2_score(y_test, y_pred_poly))

# Get the best polynomial degree
best_degree = degrees[r2_values.index(max(r2_values))]

best_degree, mse_values, r2_values
```
